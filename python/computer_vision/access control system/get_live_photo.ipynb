```python
import cv2
import numpy as np
import time
import cvlib as cv
from cvlib.object_detection import draw_bbox
import matplotlib.pyplot as plt
```

```python
def video_demo():
    capture = cv2.VideoCapture(0)
    while(True):
        ret,frame = capture.read()
        frame = cv2.flip(frame,1)
        cv2.imshow("video",frame)

        
        c = cv2.waitKey(50)
        if(c == 27):
            cv2.destroyAllWindows()
            capture.release() 
            break
    
        
```

```python
video_demo()
```

```python
def video_demo_1(capture):
    while(True):
        ret,frame = capture.read()
        frame = cv2.flip(frame,1)
        cv2.imshow("video",frame)

        
        c = cv2.waitKey(50)
        if(c == 27):
            success, img = capture.read()
            _, camphoto = cv2.imencode('.jpg', img)
            return camphoto

```

```python
capture = cv2.VideoCapture(0)
camphoto = video_demo_1(capture)

cv2.destroyAllWindows()
capture.release()

camphoto = cv2.imdecode(camphoto, cv2.IMREAD_UNCHANGED)
plt.imshow(camphoto[:,:,::-1])
```

```python
bbox ,labels ,conf = cv.detect_common_objects(camphoto)
im2 = draw_bbox(camphoto,bbox,labels,conf)
plt.imshow(im2[:,:,::-1])
```
