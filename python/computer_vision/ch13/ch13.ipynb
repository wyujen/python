```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
from imutils.object_detection import non_max_suppression
```

```python
imPath = r'p4.jpg'
im = cv2.imread(imPath)
```

```python
plt.imshow(im[:,:,::-1])
```

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
(r,w) = hog.detectMultiScale(im,winStride =(2,2),padding = (16,16),scale = 0.2)
```

```python
r
```

```python
rects = np.array([[x,y,w+h,y+h] for (x,y,w,h) in r])
#for (x,y,w,h) in r:
#    rect = np.array ([[x,y]])
```

```python
rects_NMS = non_max_suppression(rects,probs=None,overlapThresh=0.6)
```

```python
for(x1,y1,x2,y2) in rects_NMS:
    cv2.rectangle(im,(x1,y1),(x2,y2),(0,0,255),2)
```
