```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
figure , axes = plt.subplots()
draw_circle = plt.Circle((0, 0), 1, color='r',fill = False)
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)
axes.add_artist(draw_circle)
n = 0
p_x = []
p_y = []
p = []
m = 0
g = 0
while n<10000:
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    p_x.append(x)
    p_y.append(y)

    c = x*x+y*y
    n = n + 1
    if c <= 1 :
        m = m+1
    g = (4*m)/n
    p.append(g)
    
plt.scatter(p_x, p_y,c = 'black', s = 0.1 )
plt.show()
```

```python
plt.axhline(y = np.pi,color = 'black',linestyle = '--')
plt.plot(p,alpha = 0.7)
plt.show()
```
