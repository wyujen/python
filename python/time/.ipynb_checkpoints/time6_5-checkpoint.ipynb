```python
import datetime

import matplotlib 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import fft
from mpl_toolkits.mplot3d import Axes3D
import scipy
from scipy import stats
plt.style.use("C:/Users/w/Desktop/time/data/d4sci.mplstyle")
```

```python
t_max=4
sample_freq=250
N=t_max*sample_freq
t=np.linspace(0,t_max,N)
amp=np.array([1,.3,.1])
freq=np.array([1,2,10])
```

```python
def build_signal(amp,freq,t):
    total=np.zeros(len(t))
    components=[]
    n_freq=len(freq)
    
    for i in range(n_freq):
        current=amp[i]*np.cos(2*np.pi*freq[i]*t)
        total+=current
        
        components.append(current)
    return total,components          
```

```python
total,components=build_signal(amp,freq,t)
```

```python
for i ,comp in enumerate(components):
    plt.plot(t,comp,label="f=%g" % freq[i],lw=1)
plt.plot(t,total,label="total")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
```
